```python
!pip install -q spacy
!python -m spacy download en_core_web_sm > /dev/null

import spacy
from spacy import displacy
from IPython.display import HTML

nlp = spacy.load("en_core_web_sm")

text = "Apple is looking at buying U.K. startup for $1 billion. Elon Musk founded SpaceX in 2002."

doc = nlp(text)
```

```python
print("=== Part of Speech Tagging ===")
for token in doc:
    print(f"{token.text:<12}  POS: {token.pos_:<10}  Dep: {token.dep_:<10}  Head: {token.head.text}")

print("\n=== Named Entities ===")
for ent in doc.ents:
    print(f"{ent.text:<25} | {ent.label_}")
```

```python
HTML(displacy.render(doc, style="dep", jupyter=True))

HTML(displacy.render(doc, style="ent", jupyter=True))
```
