```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 10, 100)
y_true = 2 * X + 1 + np.random.randn(100) * 0.5 
w = np.random.randn()
b = np.random.randn()
lr = 0.01
epochs = 200
```

```python
losses = []

for i in range(epochs):
    y_pred = w * X + b

    loss = np.mean((y_true - y_pred) ** 2)
    losses.append(loss)

    dw = -2 * np.mean(X * (y_true - y_pred))
    db = -2 * np.mean(y_true - y_pred)

    w -= lr * dw
    b -= lr * db

    if i % 20 == 0:
        print(f"Epoch {i:3d} | Loss: {loss:.4f} | w: {w:.3f} | b: {b:.3f}")
```

```python
plt.figure(figsize=(7,4))
plt.scatter(X, y_true, label="True Data", color='blue')
plt.plot(X, w * X + b, color='red', label="Learned Line")
plt.title("Single Neuron Learning y = 2x + 1")
plt.xlabel("X"); plt.ylabel("y")
plt.legend(); plt.show()
```

```python
plt.plot(losses)
plt.title("Loss Reduction Over Epochs")
plt.xlabel("Epochs")
plt.ylabel("MSE Loss")
plt.show()
```
